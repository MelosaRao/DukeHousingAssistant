This function below extracts database by using beautiful soup and crawling over the student housing site. The code was run on google colab and is added here for reference. We use all the url extracted  below in rag.py to generate faiss_db_housing database

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

start_url = "https://students.duke.edu/living/housing/"
visited_urls = set()

def crawl(url):
    if url in visited_urls:
        return

    print("Visiting:", url)
    visited_urls.add(url)

    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        for link in soup.find_all("a", href=True):
            next_url = urljoin(url, link["href"])
            # Stay inside the domain
            if "students.duke.edu/living/housing" in next_url:
                crawl(next_url)

    except Exception as e:
        print("Error:", e)

crawl(start_url)

print(f"\nTotal pages crawled: {len(visited_urls)}")
print("\nURLs:")
for u in visited_urls:
    print(u)
```
